## Day 3 - Advent of Code

### Part A

```python
# open txt file and store each line in a list
with open('input3.txt', 'r') as file:
    lines = [line.strip() for line in file]
```

```python
# list to store total joltage for each line 
total_joltage = []

# for each line, convert string to list of chars
for line in lines:

    largest = -1
    second_largest = -1
    processed_line = list(line)  
    
    # iterates the list and stores the largest number and it's index
    for i, letter in enumerate(processed_line[:-1]):
        letter = int(letter)
        if letter > largest:
            largest = letter            
            largest_index = i  
    
    # iterates the list again from the index of the previous largest
    # number onwards, stores the next largest number and it's index
    for letter in processed_line[largest_index+1:]:
        new_letter = int(letter)
        if new_letter > second_largest:
            second_largest = new_letter

    # string combining largest and next largest digits found
    line_joltage = str(largest) + str(second_largest)

    # convert string to int and add to list of joltages
    total_joltage.append(int(line_joltage))
```

```python
# sum of joltage for each line
print(sum(total_joltage))
```

### Part B

```python
# open txt file and store each line in a list
with open('input3.txt', 'r') as file:
    lines = [line.strip() for line in file]
```

```python
# list to store total joltage for each line 
total_joltage = []

# for each line, convert string to list of chars
for line in lines:

    largest_numbers = []
    largest = -1
    processed_line = list(line)  
    endpoint = 11
    y = 11
    largest_index = 0

    # repeat y+1 times to search entire line
    for x in range(y+1):  
        
        # search slices from largest_index to -endpoint, find position of current largest 
        if endpoint != 0:
            for i, letter in enumerate(processed_line[largest_index:-endpoint]):
                letter = int(letter)
                if letter > largest:
                    largest = letter            
                    current_largest = i + 1
        
        # if endpoint is 0, search slices from largest_index to end, find position of current largest        
        else:
            for i, letter in enumerate(processed_line[largest_index:]):
                letter = int(letter)
                if letter > largest:
                    largest = letter            
                    current_largest = i + 1

        # add largest to list, update starting index, reset largest
        # and decrease endpoint to expand slice for next iteration
        largest_index = largest_index + current_largest
        largest_numbers.append(largest)
        largest = -1
        endpoint = endpoint - 1

    
    # empty string to contain line joltage
    line_joltage = ""

    # convert each largest number to a string and add to line_joltage string
    for n in largest_numbers:
        num = str(n)
        line_joltage += num
    
    # convert string to int and add to list of joltages
    total_joltage.append(int(line_joltage))
```

```python
# sum of all line joltages
print(sum(total_joltage))
```
