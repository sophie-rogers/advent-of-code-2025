## Day 6 - Advent of Code

### Part A

```python
import math

# open txt file and store each line in a list
with open('input6.txt', 'r') as file:
    rows = [line.strip().split() for line in file]

# creates columns by zipping rows by index
columns = list(zip(*rows))
```

```python
total = 0 

# check last index in list to get operator
for column in columns:
    values = [int(val) for val in column[:-1]]  # convert to ints
    operator = column[-1]

    if operator == "*":
        result = math.prod(values)
    elif operator == "+":
        result = sum(values)

    total += result

print(total)
```

### Part B

```python
import math

# open txt file and store each line in a list, stripping the newline character
with open('input6.txt', 'r') as file:
    rows = [line.strip("\n") for line in file]
```

```python
# create columns by treating each row as a list of characters
columns = list(rows)
indexes = []

# find the indices in the last row that are not a space
for index, char in enumerate(columns[-1]):
    if char != " ":
        indexes.append(index)
```

```python
formatted = []

# insert '|' before each relevant index in each row (except first index) to split later
for row in columns:
    tmp_row = list(row)
    for index in indexes:
        if index != 0:
            tmp_row[index-1] = "|"
    # split the row into separate elements using '|'
    formatted.append("".join(tmp_row).split("|"))

# extract operations from last row, remove whitespace
operations = [x.strip() for x in formatted[-1]]
# remove last row containing operations
formatted = formatted[:-1]

rotated = []
output_nums = []
```

```python
# process columns from right to left
for column in range(len(operations)-1, -1, -1):
    column_list = []
    aligned_list = []
    
    # gather all elements in the current column
    for row in formatted:
        column_list.append(row[column])
        aligned_list.append([])
        
    # find length of longest number in column
    row_length = max([len(x) for x in column_list])
    
    # align digits from bottom to top
    for index, item in enumerate(column_list):
        for x in range(row_length-1, -1, -1):
            aligned_list[x].append(item[x])

    # join digits and convert to ints, ignoring empty lists
    out = [int("".join(x).strip()) for x in aligned_list if x != []]
    output_nums.append(out)

calculated_nums = []
```

```python
# reverse operations to match the order of output numbers
operations.reverse()

# calculate vals for each column
for index, row in enumerate(output_nums):
    
    out_val = 1
    tmp = row
    tmp.reverse()
    
    if operations[index] == "+":
        # add all numbers in the column
        for item in tmp:
            out_val = out_val + item
        # subtract the initial 1 added at the start
        out_val = out_val - 1
    
    if operations[index] == "*":
        # multiply all numbers in the column
        for item in tmp:
            out_val = out_val * item
    
    # store the calculated result
    calculated_nums.append(out_val)
```

```python
# print the sum of all calculated numbers
print(sum(calculated_nums))
```
