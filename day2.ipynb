## Day 2 - Advent of Code

### Part A

```python
from itertools import islice
```

```python
# open txt file, split into tuples of min and max
with open('input2.txt', 'r') as f:
    pairs = [
        tuple(int(n) for n in element.split('-'))
        for line in f
        for element in line.strip().split(',')
        if line.strip()
    ]
```

```python
# list for invalid ids
invalid_list = []

# for each tuple in pairs generate all numbers between min and max
for min, max in pairs:
    current_list = list(range(min, max + 1))

    # iterate through all numbers in current range and convert to strings
    for x in current_list:
        whole_num = str(x)

        # split string in two halves
        half1 = ''.join(islice(whole_num, None, len(whole_num)//2 + len(whole_num)%2))
        half2 = ''.join(islice(whole_num, len(whole_num)//2 + len(whole_num)%2, None))

        # if both halves equal, convert back to int and add to list of invalid numbers
        if half1 == half2:
            invalid_num = int(whole_num)
            invalid_list.append(invalid_num)
```

```python
print(sum(invalid_list))
```

### Part B

```python
# open txt file, split into tuples of min and max
with open('input2.txt', 'r') as f:
    pairs = [
        tuple(int(n) for n in element.split('-'))
        for line in f
        for element in line.strip().split(',')
        if line.strip()
    ]
```

```python
# list for invalid ids
invalid_list = []

# for each tuple in pairs generate all numbers between min and max
for min, max in pairs:
    current_list = list(range(min, max + 1))

    # iterate through range, convert each to string, get length of string
    for x in current_list:
        whole_num = str(x)
        length = len(whole_num)

        # check all slice lengths from 1 to id length - 1
        for z in range(1, length):
            # check if slice fits into id with no remainder
            if length % z != 0:  
                continue 
            # take first z digits as slice   
            slice = whole_num[:z] 
            # how many slices would fit in the id
            repeats = length // z

            # if repeating slice == id, add to invalid list
            if slice * repeats == whole_num:
                invalid_list.append(x)
                break
```

```python
print(sum(invalid_list))
```
