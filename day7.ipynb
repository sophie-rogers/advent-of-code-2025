## Day 7 - Advent of Code

### Part A

```python
# read in file to 2d array
with open("input7.txt", "r") as f:
    grid = [list(line.strip()) for line in f if line.strip()]
```

```python
# find starting point
start_row, start_col = None, None
for row_index, row in enumerate(grid):
    for column_index, val in enumerate(row):
        if val == "S":
            start_row, start_col = row_index, column_index
            break
```

```python
# active beams list stored as row and column tuple
beams = [(start_row, start_col)]
seen_beams = set()
arrow_count = 0
rows = len(grid)
cols = len(grid[0])
```

```python
while beams:
    new_beams = []

    for row_index, column_index in beams:

        # if processed, skip
        if (row_index, column_index) in seen_beams:
            continue
        seen_beams.add((row_index, column_index))

        # move beam down one row
        below_row = row_index + 1

        # check bounds
        if below_row >= rows:
            continue

        current = grid[row_index][column_index]
        below = grid[below_row][column_index]

        if below == ".":
            # continue down
            new_beams.append((below_row, column_index))

        elif below == "^":
            # beam hits ^
            arrow_count += 1

            # left beam
            if column_index - 1 >= 0:
                new_beams.append((below_row, column_index - 1))
            # right beam
            if column_index + 1 < cols:
                new_beams.append((below_row, column_index + 1))

    beams = new_beams
```

```python
print("Total splits:", arrow_count)
```

### Part B

```python
# read in file to 2d array
with open("input7.txt") as f:
    grid = [list(line.strip()) for line in f if line.strip()]
```

```python
# find starting point
start_row, start_col = None, None
for r, row in enumerate(grid):
    for c, val in enumerate(row):
        if val == "S":
            start_row, start_col = r, c
            break
```

```python
rows = len(grid)
cols = len(grid[0])

# dictionary storing active position as column and row index, and timeline count
active = {(start_row, start_col): 1}

timelines = 0
step = 0
```

```python
while active:
    step += 1
    new_active = {}
    
    for (row_index, col_index), count in active.items():

        below_row = row_index + 1

        # leaves the end of grid array, counts a timeline
        if below_row >= rows:
            timelines += count
            continue

        cell_below = grid[below_row][col_index]

        if cell_below == ".":
            # continue downward in grid
            new_active[(below_row, col_index)] = new_active.get((below_row, col_index), 0) + count

        elif cell_below == "^":
            # split left
            if col_index - 1 >= 0:
                new_active[(below_row, col_index - 1)] = new_active.get((below_row, col_index - 1), 0) + count
            # split right
            if col_index + 1 < cols:
                new_active[(below_row, col_index + 1)] = new_active.get((below_row, col_index + 1), 0) + count

    active = new_active
```

```python
print("Total timelines:", timelines)
```
