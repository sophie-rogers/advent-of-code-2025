## Day 5 - Advent of Code

### Part A

```python
# read in file, split into ranges and ids
with open("input5.txt", "r") as file:
    raw_ranges, raw_values = file.read().split("\n\n")
```

```python
# initialise empty list for fresh ingredient ranges
ranges = []

# split each row into tuples of min and max for range
for row in raw_ranges.split("\n"):
    r_row = row.strip().split("-")
    ranges.append([int(r_row[0]), int(r_row[1])])
```

```python
# initialise empty list for available ingredient ids
available = []

for row in raw_values.strip().split("\n"):
    available.append(int(row.rstrip()))
```

```python
# initialise lists for fresh and spoiled ids, and values in ranges
fresh = []
spoiled = []
value = []

# fill each range in ranges list, if the id is in range, fresh, otherwise spoiled, added to lists accordingly
for r in ranges:
    
    value = range(r[0],r[1]+1)

    for a in available:
        if a in value:
            fresh.append(a)
        else:
            spoiled.append(a)
```

```python
print("Number of fresh ingredient ID's:", len(set(fresh)))
```

### Part B

```python
# read in file, split into ranges and ids
with open("input5.txt", "r") as file:
    raw_ranges, raw_values = file.read().split("\n\n")
```

```python
# initialise empty list for fresh ingredient ranges
ranges = []

# split each row into tuples of min and max for range
for row in raw_ranges.split("\n"):
    r_row = row.strip().split("-")
    ranges.append([int(r_row[0]), int(r_row[1])])
```

```python
# sort ranges from previous, initialise overlap range to 0
sorted_ranges = sorted(ranges)
overlap_range = 0

# initialize data1 to the first range in ranges
data1 = {"min": int(sorted_ranges[0][0]),
         "max": int(sorted_ranges[0][1])}

# for each range for the length of ranges, skipping the first, store the min and max
for i in range(1, len(sorted_ranges)):
    data2 = {"min": int(sorted_ranges[i][0]), 
             "max": int(sorted_ranges[i][1])}

    # if this min is higher than the previous max, merge the overlapping ranges
    if data2["min"] <= data1["max"]:
        data1["max"] = max(data1["max"], data2["max"])
    else:
        # otherwise count the previous merged range and start a new one beginning with current data2
        overlap_range += data1["max"] - data1["min"] + 1
        data1 = {"min": data2["min"], "max": data2["max"]}

# count the final merged range (last data1)
overlap_range += data1["max"] - data1["min"] + 1

# final total
print("Number of fresh ingredient ID's:", overlap_range)
```
