## Day 4 - Advent of Code

### Part A

```python
# read in file to 2d array
with open("input4.txt", "r") as f:
    grid = [list(line.strip()) for line in f if line.strip()]
```

```python
# sets valid rolls to zero and searches each row for an @
# if found, checks all adjacent positions for an @, adds them to roll count
# if roll count less than 4, total valid rolls increments 

valid_rolls = 0

for row_index, row in enumerate(grid):
    for column_index, val in enumerate(row):

        roll_count = 0

        # current
        if val == '@':

            # above
            if row_index > 0:
                current = grid[row_index - 1][column_index]
                if current == '@':
                    roll_count += 1

            # diagonal top left
            if row_index > 0 and column_index > 0:
                current = grid[row_index - 1][column_index - 1]
                if current == '@':
                    roll_count += 1

            # diagonal top right
            if row_index > 0 and column_index < len(row) - 1:
                current = grid[row_index - 1][column_index + 1]
                if current == '@':
                    roll_count += 1

            # below
            if row_index < len(grid) - 1:
                current = grid[row_index + 1][column_index]
                if current == '@':
                    roll_count += 1

            # diagonal bottom left
            if row_index < len(grid) - 1 and column_index > 0:
                current = grid[row_index + 1][column_index - 1]
                if current == '@':
                    roll_count += 1

            # diagonal bottom right
            if row_index < len(grid) - 1 and column_index < len(row) - 1:
                current = grid[row_index + 1][column_index + 1]
                if current == '@':
                    roll_count += 1

            # left
            if column_index > 0:
                current = grid[row_index][column_index - 1]
                if current == '@':
                    roll_count += 1

            # right
            if column_index < len(row) - 1:
                current = grid[row_index][column_index + 1]
                if current == '@':
                    roll_count += 1

            # if less than 4 adjacent rolls, roll removed
            if roll_count < 4:
                valid_rolls += 1
      
print("Total valid rolls:", valid_rolls)
```

### Part B

```python
# read in file to 2d array
with open("input4.txt", "r") as f:
    grid = [list(line.strip()) for line in f if line.strip()]
```

```python
# total rolls removed
total_removed = 0
removed_this_round = 1

# continue looping until no more can be removed in this round
while removed_this_round > 0:

    removed_this_round = 0
    to_remove = []

    # loop over each row and column
    for row_index, row in enumerate(grid):
        for col_index, val in enumerate(row):

            # if current contains @
            if val != '@':
                continue

            roll_count = 0

            # above
            if row_index > 0 and grid[row_index - 1][col_index] == '@':
                roll_count += 1
            # diagonal top left
            if row_index > 0 and col_index > 0 and grid[row_index - 1][col_index - 1] == '@':
                roll_count += 1
            # diagonal top right
            if row_index > 0 and col_index < len(row) - 1 and grid[row_index - 1][col_index + 1] == '@':
                roll_count += 1
            # below
            if row_index < len(grid) - 1 and grid[row_index + 1][col_index] == '@':
                roll_count += 1
            # diagonal bottom left
            if row_index < len(grid) - 1 and col_index > 0 and grid[row_index + 1][col_index - 1] == '@':
                roll_count += 1
            # diagonal bottom right
            if row_index < len(grid) - 1 and col_index < len(row) - 1 and grid[row_index + 1][col_index + 1] == '@':
                roll_count += 1
            # left
            if col_index > 0 and grid[row_index][col_index - 1] == '@':
                roll_count += 1
            # right
            if col_index < len(row) - 1 and grid[row_index][col_index + 1] == '@':
                roll_count += 1

            # if fewer than 4 adjacent rolls, set roll to be removed
            if roll_count < 4:
                to_remove.append((row_index, col_index))

    # remove all rolls at once
    for r, c in to_remove:
        grid[r][c] = 'x'
        removed_this_round += 1

    total_removed += removed_this_round

print("Total rolls removed:", total_removed)
```
