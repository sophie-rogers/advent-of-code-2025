## Day 1 - Advent of Code

### Part A

```python
# set current position to 50
current = 50 
count = 0
```

```python
# open txt file, split into pairs of direction and number of turns
with open('input.txt', 'r') as f:
    pairs = [
        tuple(line.strip().split(':', 1))
        for line in f
        if ':' in line and line.strip()
    ]
```

```python
# iterates through each direction and number of steps in txt file
for direction, value in pairs:
    value = int(value)   

    # if direction L, move left by value steps in one go.
    if direction == 'L':
        current = (current - value) % 100
    # else if direction R, move left by value steps in one go.
    elif direction == 'R':
        current = (current + value) % 100
    # increment count if dial on 0 at end of move
    if current == 0:
        count += 1
```

```python
print("Final position:", current)
print("Times dial ended on 0:", count)
```

### Part B

```python
# set current position to 50
current = 50 
count = 0
```

```python
# open txt file, split into pairs of direction and number of turns
with open('input.txt', 'r') as f:
    pairs = [
        tuple(line.strip().split(':', 1))
        for line in f
        if ':' in line and line.strip()
    ]
```

```python
# iterates through each direction and number of steps in txt file
for direction, value in pairs:
    value = int(value)

    # moving left one step at a time, counter incrementing when 0 is passed
    if direction == 'L':
        for i in range(value):
            current = (current - 1) % 100
            if current == 0:
                count += 1
    
    # moving right one step at a time, counter incrementing when 0 is passed        
    elif direction == 'R':
        for i in range(value):
            current = (current + 1) % 100
            if current == 0:
                count += 1
```

```python
print("Final position:", current)
print("Times dial passed 0:", count)
```
